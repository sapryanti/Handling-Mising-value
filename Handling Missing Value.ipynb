# Beberapa cara mengisi data kosong, diantaranya : 


# 1. Menggunakan Mean

```python
data_csv = pd.read_csv("dataset_airbnb.csv")

print("===========================================")
print("Dataset yang masih terdapat nilai kosong ! :")
print(data_csv.head(10))
```

```python
print("===========================================")
print("Dataset yang sudah tidak terdapat nilai kosong ! :")
data_csv = data_csv.mean().fillna(data_csv.mean())
print(data_csv.head(10))
```

# 2. Menggunakan Median

```python
data_csv = pd.read_csv("dataset_airbnb.csv")
print(data_csv.median()) #untuk mengetahui median pada data yang kosong
```

```python
data_csv.fillna(data_csv.median())
print(data_csv.head(10))
```
